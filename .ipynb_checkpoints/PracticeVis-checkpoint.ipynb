```python
#set up
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
#get data
df = pd.DataFrame()
df['x'] = random.sample(range(1,100),25)
df['y'] = random.sample(range(1,100),25)
```

```python
#sauce
df.head()
```

```python
#label
df.describe()
```

# Scatterplot  
+ https://web.stanford.edu/~mwaskom/software/seaborn/generated/seaborn.lmplot.html
+ https://stanford.edu/~mwaskom/software/seaborn/tutorial/regression.html 

```python
sns.lmplot('x','y', data=df, fit_reg=False)
```

# Density Plot  
+ https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.kdeplot.html
+ https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.distplot.html

```python
sns.kdeplot(df.y)
```

```python
sns.kdeplot(df.x)
```

```python
sns.kdeplot(df.x,df.y)
```

```python
sns.kdeplot(df.y,df.x)
```

```python
sns.distplot(df.x)
```

```python
sns.distplot(df.y)
```

# Histogram  
+ http://matplotlib.org/api/pyplot_api.html 
+ https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.rugplot.html

```python
#at this point I am noticing the output vs the vis
#also notice the bars arent multicolor as they are in source

plt.hist(df.x,alpha=.3)
sns.rugplot(df.x);
```

# Boxplot  
+ https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.boxplot.html 

```python
sns.boxplot([df.y, df.x])
# okay this vis differs significantly from the tutorial
```

# Violin 
+ https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.violinplot.html

```python
sns.violinplot([df.x, df.y])
# "at 0x#######" differs from tut
```

# Heatmap  
+ https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.heatmap.html

```python
sns.heatmap([df.y, df.x], annot=True, fmt="d")
#this seems true to source
```

# Clustermap  
+ https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.clustermap.html

```python
sns.clustermap(df)
```
